# CPU Profiling of MATLAB/Python Code
> The following graphs show the comparison between the MATLAB and Python Code.
> The python code has not been optimized with JAX to run on GPUs. The code only uses NumPy.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
matlab_results = [
    48.6272,
    49.6028,
    49.7682,
    48.3747,
    50.2922,
    50.1938,
    49.3859,
    50.5749,
    48.1928,
    50.4758
]
```

```python
plt.bar(list(range(1, 10 + 1)), matlab_results)
plt.title('MatLab Code Runtimes')
plt.xlabel('Trial #')
plt.ylabel('Time (in seconds)')

print("Average:", np.mean(matlab_results))
```

```python
python_results = [4.64580325, 
                  3.9366938750000005, 
                  3.9325037920000003, 
                  3.9140571249999994, 
                  3.92216625, 
                  3.9333744579999994, 
                  3.930374208, 
                  3.924881249999995, 
                  3.921995249999995, 
                  3.9306759999999983]
```

```python
plt.bar(list(range(1, 10 + 1)), python_results)
plt.title('Python Code Runtimes')
plt.xlabel('Trial #')
plt.ylabel('Time (in seconds)')

print("Average:", np.mean(python_results))
```

```python
plt.title('Comparison between MATLAB/Python')
plt.xlabel('Trial #')
plt.ylabel('Time (in seconds)')

# Plot the 2 bars together.
width = 0.4

plt.bar( np.array(list(range(1, 10 + 1))) - 0.2, matlab_results, width )
plt.bar( np.array(list(range(1, 10 + 1))) + 0.2, python_results, width )

plt.legend(['MATLAB', 'Python (Numpy)'])
```
